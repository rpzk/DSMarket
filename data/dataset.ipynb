```python
import pandas as pd
import tarfile
import io
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de visualização
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
```

## Estrutura dos Dados

```python
# Função para extrair um único arquivo .tar.xz e ler o CSV
def load_data_from_tarxz(tarxz_path, filename):
    with tarfile.open(tarxz_path, "r:xz") as tar:
        file = tar.extractfile(filename)
        dataframe = pd.read_csv(io.BytesIO(file.read()))
    return dataframe

# Caminhos para os arquivos .tar.xz e os nomes dos arquivos CSV dentro de cada um
tarxz_paths = {
    'item_prices': 'item_prices.tar.xz',
    'item_sales': 'item_sales.tar.xz',
}

filenames = {
    'item_prices': 'item_prices.csv',
    'item_sales': 'item_sales.csv',
    'daily_calendar': 'daily_calendar_with_events.csv'
}

# Carregar os dados
item_prices = load_data_from_tarxz(tarxz_paths['item_prices'], filenames['item_prices'])
item_sales = load_data_from_tarxz(tarxz_paths['item_sales'], filenames['item_sales'])
daily_calendar = pd.read_csv(filenames['daily_calendar'])
# Visualizar as primeiras linhas de cada DataFrame
print(item_prices.head())
print(item_sales.head())
print(daily_calendar.head())
```

```python
print(item_prices.info())
print(item_prices.describe())
```

```python
print(item_sales.info())
print(item_sales.describe())
```

```python
print(daily_calendar.info())
print(daily_calendar.describe())
```

## Limpeza e Preparação dos Dados

```python
# Preencher valores nulos nas colunas numéricas
item_prices.fillna(item_prices.select_dtypes(include='number').median(), inplace=True)
item_sales.fillna(item_sales.select_dtypes(include='number').median(), inplace=True)

# Preencher valores nulos em colunas categóricas com um valor específico ou usando um método adequado
item_prices.fillna({'item': 'Unknown', 'category': 'Unknown', 'store_code': 'Unknown', 'yearweek': 0}, inplace=True)
item_sales.fillna({'id': 'Unknown', 'item': 'Unknown', 'category': 'Unknown', 'department': 'Unknown', 'store': 'Unknown', 'store_code': 'Unknown', 'region': 'Unknown'}, inplace=True)

# Preencher valores nulos no calendário
daily_calendar.fillna('No Event', inplace=True)
```

## Análise Exploratória

### Preços

```python
plt.figure(figsize=(10, 6))
sns.histplot(item_prices['sell_price'], bins=50, kde=True)
plt.title('Distribuição de Preços de Venda')
plt.xlabel('Preço de Venda')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=item_prices, x='yearweek', y='sell_price', hue='item')
plt.title('Evolução do Preço de Venda por Item ao Longo do Tempo')
plt.xlabel('Ano/Semana')
plt.ylabel('Preço de Venda')
plt.show()
```

### Vendas

```python
item_sales['total_sales'] = item_sales.iloc[:, 7:].sum(axis=1)
plt.figure(figsize=(10, 6))
sns.barplot(data=item_sales, x='item', y='total_sales')
plt.title('Vendas Totais por Item')
plt.xlabel('Item')
plt.ylabel('Vendas Totais')
plt.xticks(rotation=90)
plt.show()
```

### Eventos por dia da semana

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=daily_calendar, x='weekday', hue='event')
plt.title('Distribuição de Eventos por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Contagem de Eventos')
plt.show()
```
