```python
from google.oauth2 import service_account
from google.cloud import bigquery
import os

# Caminho para o arquivo JSON com as credenciais da Service Account
key_path = '/home/rp/Desenvolvimento/DSMarket/eda/tfm-sa.json'  # Updated path

# Cria as credenciais da Service Account
credentials = service_account.Credentials.from_service_account_file(key_path)

# Inicializa o cliente BigQuery
client = bigquery.Client(credentials=credentials, project=credentials.project_id)

# Nome do dataset e tabela onde o CSV será carregado
dataset_id = 'TFM'
table_id = 'ds_market'  # Substitua pelo nome da tabela que deseja criar

# Caminho para o arquivo CSV
csv_file_path = '../data/ds_market.csv'

# Configurações para o job de upload
job_config = bigquery.LoadJobConfig(
    source_format=bigquery.SourceFormat.CSV,
    skip_leading_rows=1,  # Pula o cabeçalho, caso necessário
    autodetect=True,  # Detecta o esquema automaticamente
)

# Realiza o upload do CSV para o BigQuery
with open(csv_file_path, "rb") as source_file:
    job = client.load_table_from_file(source_file, f"{dataset_id}.{table_id}", job_config=job_config)

# Espera o job ser concluído
job.result()

# Verifica se o upload foi bem-sucedido
table = client.get_table(f"{dataset_id}.{table_id}")
print(f"Carregado {table.num_rows} linhas para a tabela {table_id}.")
```
