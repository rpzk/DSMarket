```python
import pandas as pd
import glob
import os
from google.oauth2 import service_account
from pandas_gbq import to_gbq

# Caminho para o diretório com seus arquivos Parquet
parquet_dir = '../data/ds_market/'

# Obter a lista de todos os arquivos Parquet
parquet_files = glob.glob(os.path.join(parquet_dir, 'part.*.parquet'))

# Carregar e concatenar todos os arquivos Parquet em um único DataFrame
df_list = [pd.read_parquet(f) for f in parquet_files]
df = pd.concat(df_list, ignore_index=True)
```

```python
# Caminho para o arquivo JSON com as credenciais da Service Account
key_path = 'tfm-sa.json'

# Cria as credenciais da Service Account
credentials = service_account.Credentials.from_service_account_file(key_path)

# IDs do projeto e dataset
project_id = 'perseverance-332400'
dataset_id = 'TFM'
table_id = 'ds_market'  # Nome da tabela que deseja criar ou substituir

# Fazer o upload para o BigQuery
to_gbq(
    df,
    f"{dataset_id}.{table_id}",
    project_id=project_id,
    if_exists='replace',  # Use 'append' se quiser adicionar dados a uma tabela existente
    credentials=credentials
)

print(f"Carregado {len(df)} linhas para a tabela {table_id}.")
```
